```python
data = pd.read_csv('sharma-kohli.csv')
data
plt.plot(data['index'], data['V Kohli'], color='yellow',label='virat',marker='o')
plt.plot(data['index'], data['RG Sharma'], color='blue', label='rohit',marker='o')
plt.xlabel('Seasons')
plt.ylabel('Scores')
plt.title('Rohit vs Virat score comparision')
plt.legend()
plt.grid()
plt.show()
```

```python
# Multiple Bar charts
df = pd.read_csv('batsman_season_record.csv')
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2, color='#17BECF',label='2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, color='#9467BD',label='2016')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2, color='#FF7F0E',label='2017')
plt.legend()
plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017', color='#0072B2')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016', color='#E69F00')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015', color='#009E73')

plt.legend()
plt.show()
```
