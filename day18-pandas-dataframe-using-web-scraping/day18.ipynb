# creating dataframe for all the pages

```python
dfs = []  # ‚Üê list to collect each page‚Äôs DataFrame

for j in range(1, 504):
    webpage = requests.get(
        f'https://www.ambitionbox.com/list-of-companies?page={j}'
    ).text
    soup = BeautifulSoup(webpage, 'lxml')
    company = soup.find_all('div', class_='company-content-wrapper')

    # temporary lists for this page
    name = []
    rating = []
    reviews = []
    ctype = []
    hq = []
    how_old = []
    no_of_employee = []

    for i in company:
        # each of these try/except blocks stays the same
        try:
            name.append(i.find('h2').text.strip())
        except:
            name.append(np.nan)

        try:
            rating.append(i.find('p', class_='rating').text.strip())
        except:
            rating.append(np.nan)

        try:
            reviews.append(i.find('a', class_='review-count').text.strip())
        except:
            reviews.append(np.nan)

        try:
            ctype.append(i.find_all('p', class_='infoEntity')[0].text.strip())
        except:
            ctype.append(np.nan)

        try:
            hq.append(i.find_all('p', class_='infoEntity')[1].text.strip())
        except:
            hq.append(np.nan)

        try:
            how_old.append(i.find_all('p', class_='infoEntity')[2].text.strip())
        except:
            how_old.append(np.nan)

        try:
            no_of_employee.append(i.find_all('p', class_='infoEntity')[3].text.strip())
        except:
            no_of_employee.append(np.nan)

    # build the DataFrame for this page
    df = pd.DataFrame({
        'name': name,
        'rating': rating,
        'reviews': reviews,
        'company_type': ctype,
        'Head_Quarters': hq,
        'Company_Age': how_old,
        'No_of_Employee': no_of_employee,
    })

    dfs.append(df)  # ‚Üê append the page-DF to our list

# once the loop is done, merge them all in one go
final = pd.concat(dfs, ignore_index=True)
```
